Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose you have collected data on the amount of time students spend studying for an exam and their final exam scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```python
import pandas as pd
# Sample data
study_time=[2, 3, 5, 7, 9]
exam_scores=[50, 55, 65, 75, 85]
#Sample Data Convert into Pandas DataFrame
df=pd.DataFrame({'Study Time': study_time, 'Exam Scores': exam_scores})
#Calculate the Pearson correlation coefficient between these two variables and interpret the result.
df.corr(method='pearson')
```

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables. Suppose you have collected data on the amount of sleep individuals get each night and their overall job satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

```python
import pandas as pd
# Sample data
sleep_hrs=[4, 5, 6, 7, 8]
job_satisfaction=[3, 4, 7, 8, 9]
df2=pd.DataFrame({'Sleep Hrs':sleep_hrs, 'Job Satisfaction': job_satisfaction})
#Calculate the Spearman's rank correlation between these two variables and interpret the result.
df2.corr(method='spearman')
```

Q3. Suppose you are conducting a study to examine the relationship between the number of hours of exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation between these two variables and compare the results.

```python
import numpy as np
# Exercise hours per week (0-10)
exercise_hours=np.random.randint(0, 10, 50)
# BMI values between 18 and 35
bmi=np.random.randint(18, 30, 50)
#Sample data convert into Pandas DataFrame
df3=pd.DataFrame({'Exercise Hours': exercise_hours, 'BMI': bmi})
#Calculasting the Correlation b/w Exercise Hours and BMI via Pearson Coffiecient Correlation
df3.corr(method='pearson')
#Calculasting the Correlation b/w Exercise Hours and BMI via Spearman Coffiecient Correlation
df3.corr(method='spearman')
```

Q5. A survey was conducted to examine the relationship between age and preference for a particular brand of soft drink. The survey results are shown below:
Screenshot 2025-03-15 112403.png

```python
# Step 1: Create a DataFrame with age and soft drink preference
age_years=[25, 42, 37, 19, 31, 28]
soft_drink=['Coke', 'Pepsi', 'Mountain Dew', 'Coke', 'Pepsi', 'Coke']

df5=pd.DataFrame({'Age Years':age_years, 'Soft Drink':soft_drink})

# Step 2: Perform Data Encoding - Convert categorical values into numerical values
# Compute the mean age for each soft drink category
soft_drink_mean=df5.groupby(['Soft Drink'])['Age Years'].mean().to_dict()

# Step 3: Map the mean age to each row based on the soft drink preference
df5['Age Year Mean']=df5['Soft Drink'].map(soft_drink_mean)

# Step 4: Select only numeric columns for correlation analysis
df5=df5.select_dtypes(include='number')

# Step 5: Calculate the Spearman correlation between numeric columns
corr_matrix=df5.corr(method='spearman')

# Display the correlation matrix
print(corr_matrix)
```

 Q6. A company is interested in examining the relationship between the number of sales calls made per day and the number of sales made per week.
```python
import numpy as np
import pandas as pd

#Step1: Create Data Frame with Sales Call Per Day and Sales Call Per Week
sales_call_per_day=np.random.randint(10,50,30)
sales_call_per_week=np.random.randint(500,1000,30)

df6=pd.DataFrame({'Sales Call Per Day':sales_call_per_day, 'Sales Call Per Week':sales_call_per_week})

#Step2: Calculate the Pearson correlation coefficient between Sales Call Per Day and Sales Call Per Week
correlation_matrix=df6.corr(method='pearson')

#Step3: Display the Pearson correlation coefficient between these two variables
correlation_matrix
```
